## An accurate boundary-integral formulation for the calculation of electrostatic forces with an implicit-solvent model

### Libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler
#change plt font size\
plt.style.use('default')
plt.rcParams['lines.markersize'] = 8
plt.rcParams['font.size'] = 14
my_colors = ["red","blue","green","goldenrod"]
custom_cycler = cycler(color=my_colors)
```

## Barnase-Barstar complex

### Figure of binding energy and forces profile for Barnase-Barstar complex at several distances of separation along z-axis

#### Barnase

```python
force_terms = np.loadtxt('results_pbj/barnase-barstar_forces_to_latex_mesh8.txt',delimiter='&',skiprows=1)
plt.figure(figsize=(8,6))
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

lns1 = ax1.plot(force_terms[1:,0],(force_terms[1:,7]-force_terms[:-1,7])/(force_terms[1:,0]-force_terms[:-1,0]),label='Method 1',marker='^',mfc='none')
lns2 = ax1.plot(force_terms[:,0],force_terms[:,1]-force_terms[-1,1],label='Method 2',marker='o',mfc='none')
lns3 = ax1.plot(force_terms[:,0],force_terms[:,2]-force_terms[-1,2],label='Method 3',marker='x',mfc='none')
lns4 = ax1.plot(force_terms[:,0],force_terms[:,3]-force_terms[-1,3],label='Method 4',marker='v',mfc='none')
lns5 = ax2.plot(force_terms[:,0],force_terms[:,7],label='$\Delta G_{bind}$',marker='d',mfc='none',color='black')

# added these three lines
lns = lns1+lns2+lns3+lns4+lns5
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=4)
plt.xlim([8, 31])
ax1.set_xlabel('Distance ($\AA$)')
ax1.set_ylabel('$F_{bind}$ (kcal/mol/$\AA$)')
ax2.set_ylabel('$G_{bind}$ (kcal/mol)')
# ax1.grid()
plt.title('Barnase')
plt.savefig('figures//Barnase_formul_dist_mesh8.png',dpi=100,bbox_inches='tight')
```

## Two spheres

### Two spheres on several distances: Figure of solvation forces by several grid mesh

```python
# read results file
force_terms = np.loadtxt('results_pbj\sphere_r1q2_forces_to_latex.txt',delimiter='&',skiprows=1)
plt.figure(figsize=(8,6))
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

lns1 = ax1.plot(force_terms[:,0],np.ones(len(force_terms[:,0]))*1.9425,label='Analytical',marker='^',mfc='none')
lns2 = ax1.plot(force_terms[:,0],force_terms[:,1],label='Method 1',marker='^',mfc='none')
lns3 = ax1.plot(force_terms[:,0],force_terms[:,3],label='Method 2',marker='o',mfc='none')
lns4 = ax1.plot(force_terms[:,0],force_terms[:,2],label='Method 3',marker='x',mfc='none')
lns5 = ax1.plot(force_terms[:,0],force_terms[:,4],label='Method 4',marker='v',mfc='none')
lns6 = ax2.plot(force_terms[:,0],force_terms[:,5],label='$\Delta G_{bind}$',marker='d',mfc='none',color='black')

# added these three lines
lns = lns1+lns2+lns3+lns4+lns5+lns6
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc='best')
plt.xticks([4,8,16,32])
ax1.set_xlabel('Grid (vert/$\AA ^2$)')
ax1.set_ylabel('$F_{bind}$ (kcal/mol/$\AA$)')
ax2.set_ylabel('$G_{bind}$ (kcal/mol)')
ax1.grid()
plt.title('Sphere')
plt.savefig('figures//Sphere_formul_grid2.png',dpi=100,bbox_inches='tight')
```

### Relative force value over grid refinement

```python
# read results file
def relative_value(x):
    return np.abs((x-0.0001)-1.9425)/1.9425 #1.9425 is the analytical value, 0.0001 is the value of the force far away

force_terms = np.loadtxt(os.path.join('results_pbj','sphere_r1q2_forces_to_latex.txt'),delimiter='&',skiprows=1)
plt.figure(figsize=(8,6))
plt.rc('axes', prop_cycle=custom_cycler)
plt.plot(force_terms[:,0],relative_value(force_terms[:,1]),label='Method 1',marker='^',mfc='none')
plt.plot(force_terms[:,0],relative_value(force_terms[:,3]),label='Method 2',marker='o',mfc='none')
plt.plot(force_terms[:,0],relative_value(force_terms[:,2]),label='Method 3',marker='x',mfc='none')
plt.plot(force_terms[:,0],relative_value(force_terms[:,4]),label='Method 4',marker='v',mfc='none')
plt.plot(force_terms[:,0],0.5e-1*1/force_terms[:,0],ls='--',color='black')

# added these three lines
plt.legend(loc='best')
plt.xticks([4,8,16,32,64])
plt.xlabel('Grid (vert/$\AA ^2$)')
plt.ylabel('Relative Error (-)')
plt.yscale('log')
plt.xscale('log')
plt.grid()
plt.savefig(os.path.join('figures','Sphere_formul_grid_rel.png'),dpi=100,bbox_inches='tight')
```

```python
# read results file
def relative_value(x,term):
    if term=='Fqf':
        return np.abs(x-2.693726)/2.693726
    elif term=='Fdb':
        return np.abs(x-0.847832)/0.847832
    elif term=='Fib':
        return np.abs(x-0.016133)/0.016133

force_terms = np.loadtxt(os.path.join('results_pbj','sphere_r1q2_forces_terms_to_latex.txt'),delimiter='&',skiprows=1)
plt.figure(figsize=(8,6))
plt.rc('axes', prop_cycle=custom_cycler)
plt.plot(force_terms[:,0],relative_value(force_terms[:,1],'Fqf'),label='Fqf',marker='^',mfc='none')
plt.plot(force_terms[:,0],relative_value(force_terms[:,3],'Fdb'),label='Fdb',marker='o',mfc='none')
plt.plot(force_terms[:,0],relative_value(force_terms[:,4],'Fib'),label='Fib',marker='x',mfc='none')
plt.plot(force_terms[:,0],0.5e-1*1/force_terms[:,0],ls='--',color='black')

# added these three lines
plt.legend(loc='best')
plt.xticks([4,8,16,32,64])
plt.xlabel('Grid (vert/$\AA ^2$)')
plt.ylabel('Relative Error (-)')
plt.yscale('log')
plt.xscale('log')
plt.grid()
plt.savefig(os.path.join('figures','Sphere_terms_grid_rel.png'),dpi=100,bbox_inches='tight')
```
